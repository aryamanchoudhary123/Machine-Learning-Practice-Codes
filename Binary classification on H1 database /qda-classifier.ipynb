```python
# importing or loading the dataset
dataset = pd.read_csv('/Users/aryaman/Desktop/COLLEGE/SEMESTER 6/Predictive analysis/Lab eval Kaggle/2022-ucs654-Lab-2-Exam/h1-train_data.csv')
test  = pd.read_csv("/Users/aryaman/Desktop/COLLEGE/SEMESTER 6/Predictive analysis/Lab eval Kaggle/2022-ucs654-Lab-2-Exam/h1-test_data.csv")

# distributing the dataset into two components X and Y
X = dataset.iloc[:, 1:-1].values
y = dataset.iloc[:,-1].values
```

```python
X_train = dataset.iloc[:,1:-1]
X_test = test.iloc[:,1:]
y_train = dataset.iloc[:,-1]
X_test
```

```python
X_train = X_train.iloc[:15000,:]
y_train = y_train.iloc[:15000]
y_train
```

```python
from sklearn.linear_model import LogisticRegression 
#from sklearn import discriminant_analysis
 
classifier = LogisticRegression(penalty='l2', dual=False, tol=0.0001, 
                          C=0.01, fit_intercept=True, intercept_scaling=1.0, 
                         class_weight=None, random_state=None , solver="liblinear",max_iter=5100)

#classifier = discriminant_analysis.QuadraticDiscriminantAnalysis(priors=None, reg_param=0.5,tol=0.0001)
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
```
